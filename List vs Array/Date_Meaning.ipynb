```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import math
```

```python

df = pd.read_csv("List_Array.csv")

List = df[df["DataType"] == "List"]
Array = df[df["DataType"] == "Array"]
```

```python
List_Giga = List["Size"].sum() / 8000000000
Array_Giga = Array["Size"].sum() / 8000000000
```

```python
plt.bar(["List", "Array"], [List_Giga, Array_Giga], color='maroon', width=0.4)
plt.xlabel("Data Type")
plt.ylabel("The size is in gigabytes")
plt.title("The volume required to store {}".format(df["Factorial"].sum()))
plt.show()
```

```python
print("The Array used less space than the List {} times".format(round(List_Giga/Array_Giga)))
```

```python
plt.plot(Array['Factorial'].values, Array['Duration'])
plt.plot(List['Factorial'].values, List['Duration'])

plt.title('Duration for each operation')
plt.xlabel('Factorial Number')
plt.ylabel('Time in seconds')
plt.legend(["Array", "List"])
plt.show()
```

```python
sns.boxplot(List['Duration'])
plt.show()
```

```python
outliers = List[List['Duration'] > 0.45].index
List = List.drop(outliers)
```

```python
sns.boxplot(List['Duration'])
plt.show()
```

```python
sns.boxplot(Array['Duration'])
plt.show()
```

```python
Q1 = np.percentile(Array['Duration'], 25,
                   interpolation='midpoint')

Q3 = np.percentile(Array['Duration'], 75,
                   interpolation='midpoint')
IQR = Q3 - Q1
upper = Q3 + 1.5 * IQR
outliers = Array[Array['Duration'] > upper].index
Array = Array.drop(outliers)
```

```python
sns.boxplot(Array['Duration'])
plt.show()
```

```python
plt.plot(Array['Factorial'].values, Array['Duration'])
plt.plot(List['Factorial'].values, List['Duration'])

plt.title('Duration for each operation')
plt.xlabel('Factorial Number')
plt.ylabel('Time in seconds')
plt.legend(["Array", "List"])
plt.savefig('DurationOutlier.png')
```

```python
plt.scatter(Array['Factorial'].values, Array['Duration'])
plt.scatter(List['Factorial'].values, List['Duration'])

plt.title('Duration for each operation')
plt.xlabel('Factorial Number')
plt.ylabel('Time in seconds')
plt.legend(["Array", "List"])
plt.savefig('DurationOutlier.png')
```

```python
print("Array Mean",np.mean(Array['Duration']))
print("List Mean",np.mean(List['Duration']))
```

```python
print("The Array more performance than the List {} times".format(round(np.mean(List['Duration'])/np.mean(Array['Duration']))))
```
